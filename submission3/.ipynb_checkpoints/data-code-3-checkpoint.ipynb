### Question 1: Provide a table of mean hospital uncompensated care (in millions of dollars) by year, from 2010 through 2018. How has uncompensated care changed over time?

```r
q1_summary <- final.hcris %>%
  mutate(uncomp_care_val = tot_uncomp_care_charges - tot_uncomp_care_partial_pmts + bad_debt) %>%
  mutate(year = year(mdy(fy_end))) %>%
  group_by(year) %>%
  summarize(
    `uncomp care` = mean(uncomp_care_val / 1e6, na.rm = TRUE),
    `# hospitals` = n()
  ) %>%
  filter(year >= 2010 & year <= 2018) %>%
  as.data.frame()

q1_summary
```

Uncompensated care rose from 34.3M in 2011 to a peak of 39.2M in 2013 just before the major ACA expansion. After 2013, there was a notable decline through 2015, though more recent years show some interesting fluctuations with another spike in 2016 followed by a gradual decrease back toward 38M by 2018.

### Question 2: Plot mean uncompensated care over time separately for states that expanded Medicaid in 2014 versus states that never expanded. Drop all states that expanded after 2014. Does the graph suggest a potential treatment effect?

```r
q2_plot_data <- final_data %>%
  filter(expansion_year == 2014 | is.na(expansion_year)) %>%
  filter(fyear >= 2010 & fyear <= 2018) %>% 
  mutate(group = ifelse(is.na(expansion_year), "Never Expanded", "Expanded 2014")) %>%
  group_by(fyear, group) %>%
  summarize(mean_uncomp = mean(uncomp_millions, na.rm = TRUE), .groups = "drop")

ggplot(q2_plot_data, aes(x = fyear, y = mean_uncomp, color = group, group = group)) +
  geom_line(linewidth = 1.2) +
  geom_point(size = 2) +
  # Vertical line positioned between 2013 and 2014
  geom_vline(xintercept = 2013.5, linetype = "dashed", color = "black") +
  scale_x_continuous(breaks = 2010:2018) +
  labs(
    title = "Uncompensated Care Trends: 2014 Expanders vs. Never Expanders",
    x = "Fiscal Year",
    y = "Mean Uncompensated Care ($ Millions)",
    color = "State Group"
  ) +
  theme_minimal()
```

The graph does suggest a treatment effect because the Expanded 2014 states saw a sharp, steady decrease in uncompensated care immediately following the policy change, while the Never Expanded states saw their uncompensated care costs continue to rise.

### Question 3: Using 2012 and 2015 as your pre and post periods, present a 2x2 DD table of mean uncompensated care for expansion versus non-expansion states.

```r
q3_table <- final_data %>%
  filter(fyear %in% c(2012, 2015)) %>%
  
# Filter for 2014 expanders and never-expanders
filter(expansion_year == 2014 | is.na(expansion_year)) %>%
  filter(uncomp_millions > 0, uncomp_millions < 450) %>%
  
# Create labels and calculate means
mutate(group = ifelse(is.na(expansion_year), "Never Expanded", "Expanded 2014"),
     period = ifelse(fyear == 2012, "Pre (2012)", "Post (2015)")) %>%
group_by(group, period) %>%
summarize(mean_uncomp = mean(uncomp_millions, na.rm = TRUE), .groups = "drop") %>%
pivot_wider(names_from = period, values_from = mean_uncomp) %>%
select(group, `Pre (2012)`, `Post (2015)`)

q3_table
```

### Question 4: Briefly discuss what policies or events might explain the trends you observe. Why might we expect Medicaid expansion to affect uncompensated care?

The trends we see here, specifically the sharp drop in Expanded and the continued rise in Never Expanded uncompensated care costs, are mainly driven by the implementation of the Affordable Care Act (ACA) and the specific mechanisms of Medicaid expansion. Another factor is the Supreme Court ruling in 2012 that made expansion optional, contributing further to differences across states.

We expect a negative relationship because Medicaid acts as a financial buffer. In non-expansion states, the Never Expanded uncompensated care rises because these hospitals are still treating the same number of uninsured patients, but the cost of treating them is increasing over time due to medical inflation. In expansion states, the government takes those costs off of hospitals.

### Question 5: Estimate the effect of Medicaid expansion on hospital uncompensated care using a standard DD regression estimator, focusing only on states that expanded in 2014 versus those that never expanded.

```r
reg_dat_q5 <- final_data %>%
  # Filter only for 2014 expanders and never-expanders, but keep ALL years
  filter(expansion_year == 2014 | is.na(expansion_year)) %>%
  filter(fyear <= 2018) %>%
  filter(uncomp_millions > 0, uncomp_millions < 450) %>%
  mutate(
    expand_ever = ifelse(!is.na(expansion_year), 1, 0),
    post = ifelse(fyear >= 2014, 1, 0),
    treat = post * expand_ever
  )

model_q5 <- lm(uncomp_millions ~ post + expand_ever + treat, data = reg_dat_q5)
summary(model_q5)
```

### Question 6: Include hospital and year fixed effects in your estimates using the fixest package. Cluster your standard errors at the state level. How do your results compare to those in question 5?

```r
reg_data_q6 <- final_data %>%
filter(expansion_year == 2014 | is.na(expansion_year)) %>%
  filter(fyear <= 2018) %>%
  filter(uncomp_millions > 0, uncomp_millions < 450) %>%
  mutate(
    treat = ifelse(!is.na(expansion_year) & fyear >= 2014, 1, 0)
  )

# Run the TWFE model
model_q6 <- feols(uncomp_millions ~ treat | provider_number + fyear, 
                  data = reg_data_q6, 
                  cluster = ~state)

summary(model_q6)
```

### Q5 and Q6 Comparison

```r
library(modelsummary)

# Force both models to use same name
names(model_q5$coefficients)[names(model_q5$coefficients) == "treat"] <- "Post x Expand"
names(model_q6$coefficients) <- "Post x Expand" 

final_map <- c(
    "(Intercept)" = "(Intercept)",
    "post" = "Post 2014",
    "expand_ever" = "Expand",
    "Post x Expand" = "Post x Expand"
)

msummary(list("Standard DD" = model_q5, "TWFE" = model_q6), 
         stars = TRUE,
         coef_map = final_map,
         gof_omit = "AIC|BIC|Log.Lik|RMSE|F|Adj.|Pseudo",
         output = "markdown")
```

The estimated impact of Medicaid expansion increased when fixed effects were added (-12M to -18M). This suggests that once unique characteristics of individual hospitals are incorporated, the drop in uncompensated care is much larger than the raw averages suggested. Also, the Adjusted R squared increased from 2% to 87%. This shows that while the simple model in Q5 explained very little, the inclusion of hospital and year fixed effects in Q6 accounts for the majority of the variation in uncompensated care costs.

### Question 7: Repeat the analysis in question 6 but include all states (even those that expanded after 2014). Are your results different? If so, why?

```r
# Create data for Q7 w same filters from Q5 and Q6
q7_dat <- final_data %>%
filter(fyear <= 2018) %>%
  filter(uncomp_millions > 0, uncomp_millions < 450) %>%
  mutate(
    treat = ifelse(!is.na(expansion_year) & fyear >= expansion_year, 1, 0)
  )

# Run staggered model
model_q7 <- feols(uncomp_millions ~ treat | provider_number + fyear, 
                  data = q7_dat, 
                  cluster = ~state)

summary(model_q7)
```

### Q5-Q7 Comparison

```r
# Make sure names are identical
names(model_q5$coefficients)[names(model_q5$coefficients) == "treat"] <- "Post x Expand"
names(model_q6$coefficients) <- "Post x Expand"
names(model_q7$coefficients) <- "Post x Expand"

# Create list
full_comparison_list <- list(
  "Standard DD" = model_q5,
  "TWFE" = model_q6,
  "Staggered"   = model_q7
)

# Table
library(modelsummary)
msummary(full_comparison_list, 
         stars = TRUE,
         coef_map = c("(Intercept)" = "(Intercept)",
                      "post" = "Post 2014",
                      "expand_ever" = "Expand",
                      "Post x Expand" = "Post x Expand"),
         gof_omit = "AIC|BIC|Log.Lik|RMSE|F|Adj.|Pseudo",
         output = "markdown")
```

The addition of the staggered model in Q7 confirms that the reduction in uncompensated care is a nationwide phenomenon, not just a localized trend among early-adopting states. Even when capping at 2018 and including later adopters (increased N), the treatment effect remains statistically significant. While the staggered estimate of -14.74M is smaller than the TWFE estimate of -18.27M, this is normal in staggered designs. It reflects the averaging of early states that have had years to see results with late states that were only just beginning to see changes by 2018.

### Question 8: Provide an “event study” graph showing the effects of Medicaid expansion in each year. 

```r
event_study <- feols(uncomp_millions ~ i(fyear, expand, ref = 2013) | 
                       provider_number + fyear, 
                     cluster = ~state, 
                     data = final_data %>% filter(fyear <= 2018))

# Plot
iplot(event_study, 
      main = "Event Study: Medicaid Expansion Impact (2008-2018)",
      xlab = "Fiscal Year (With 2013 Baseline)",
      ylab = "Effect on Uncompensated Care ($M)",
      ref.line = TRUE)
```

### Question 9: Repeat part 8 but include all states that expanded after 2014. You will need to construct an “event time” variable and bin end points as discussed in class.

```r
# Event time = current year - year of expansion
event_study_data <- final_data %>%
  mutate(
    event_time = fyear - expansion_year,
    # Bin  endpoints: -5 for anything 5+ years before, 4 for 4+ years after
    event_time_binned = case_when(
      event_time <= -5 ~ -5,
      event_time >= 4  ~ 4,
      TRUE ~ event_time
    )
  ) %>%

  mutate(dist_to_expansion = factor(event_time_binned))

# Ref = -1 as baseline (year before expansion)
model_event_binned <- feols(uncomp_millions ~ i(event_time_binned, ref = -1) | 
                            provider_number + fyear, 
                            cluster = ~state, 
                            data = event_study_data)

iplot(model_event_binned, 
      main = "Event Study: Binned Event Time (All Expanders)",
      xlab = "Years Relative to Expansion",
      ylab = "Estimate and 95% CI ($ Millions)")
```

### Question 10: Summarize your findings from questions 5-9. What is the effect of Medicaid expansion on hospital uncompensated care? Do the event study graphs suggest any concerns about the parallel trends assumption? Briefly discuss one limitation of the standard TWFE estimator in the context of staggered Medicaid expansion.

- Medicaid expansion consistently reduced hospital uncompensated care costs from its implementation onward.
- Using TWFE versus DD model  increased the Adjusted $R^2$, showing that hospital-specific and year-to-year factors account for most of the cost variation.
- Event study graphs validate parallel trends assumption, as pre-expansion coefficients were flat and statistically insignificant, showing no divergence before the policy implementation.
- One limitation of the standard TWFE estimator is that it might underestimate the true effect due to the weighting issues that come up when comparing early-adopters to late-adopters. The TWFE estimator assumes the treatment effect is constant, but since Medicaid's impact tends to grow over time, comparing new expanders to old expanders can create a downward bias.
